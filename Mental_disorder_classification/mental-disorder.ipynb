```python
mapping_dict = {'Normal': 0, 'Bipolar Type-1': 1, 'Bipolar Type-2': 2, 'Depression': 3}
mental_dis['Expert Diagnose'] = mental_dis['Expert Diagnose'].map(mapping_dict).astype(int)
mental_dis.head()
```

```python
yes_and_no_columns = ['Mood Swing', 'Suicidal thoughts', 'Anorxia', 'Authority Respect', 'Try-Explanation', 'Aggressive Response', 'Ignore & Move-On', 'Nervous Break-down', 'Admit Mistakes', 'Overthinking']

for column in yes_and_no_columns:
    mental_dis[column] = mental_dis[column].map({'YES': 1, 'NO': 0}).astype(int)

mental_dis.head()
```

```python
mental_dis['Sexual Activity'].dtype
```

```python
mental_dis = mental_dis.rename(columns={'Optimisim': 'Optimism'})
mental_dis = mental_dis.rename(columns={'Anorxia': 'Anorexia'})
from_columns = ['Sexual Activity', 'Concentration', 'Optimism']
for column in from_columns:
    mental_dis[column] = mental_dis[column].astype(str).str.extract('(\d)')
    mental_dis[column] = pd.to_numeric(mental_dis[column])

mental_dis.head()
```

```python
mental_dis['Sexual Activity'].dtype
```

```python
frequency_columns = ['Sadness', 'Euphoric', 'Exhausted', 'Sleep dissorder']
for column in frequency_columns:
    mental_dis[column] = mental_dis[column].map({'Seldom': 0, 'Sometimes': 1, 'Usually': 2, 'Most-Often': 3}).astype(int)

mental_dis.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegressionCV
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report,ConfusionMatrixDisplay
```

```python
X = mental_dis.iloc[:, :-1]
y = mental_dis['Expert Diagnose']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1 ,random_state=101)
scaler = MinMaxScaler()
scaled_X_train = scaler.fit_transform(X_train)
scaled_X_test = scaler.transform(X_test)
#fit model with l1,l2,elasticnet
logCV_model = LogisticRegressionCV(solver="saga", multi_class="ovr", max_iter= 5000000,penalty= "l1",l1_ratios=np.linspace(0,1,100),cv = 10, Cs= np.logspace(1,10,100))
logCV_model.fit(scaled_X_train,y_train)
```

```python
#l1
y_pred= logCV_model.predict(scaled_X_test)
print(classification_report(y_test,y_pred))
cm = confusion_matrix(y_test,y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm,display_labels=logCV_model.classes_)
disp.plot()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import GridSearchCV

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.05, random_state = 0)

# define the parameter grid
param_grid = {
    'n_estimators': [50, 100, 200],
    'max_depth': [None, 5, 10],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

rf_classifier = RandomForestClassifier(random_state = 0)

# instantiate GridSearchCV
grid_search = GridSearchCV(estimator = rf_classifier, param_grid = param_grid, verbose = 0)

# fit the grid search to the data
grid_search.fit(X_train, y_train)

# set the best parameters
best_params = grid_search.best_params_
print("Best Parameters:", best_params)
```

```python
optimized_rf_classifier = RandomForestClassifier(**best_params, random_state = 0) # dict unpacking

# fit the new classifier on the training data
optimized_rf_classifier.fit(X_train, y_train)

# make predictions on the test and train data
y_pred_test = optimized_rf_classifier.predict(X_test)
y_pred_train = optimized_rf_classifier.predict(X_train)

# evaluate the model
# calculate accuracy on the training set
train_accuracy = accuracy_score(y_train, y_pred_train)
print(f"Training Accuracy: {train_accuracy:.2%}")

# calculate accuracy on the test set
rf_accuracy = accuracy_score(y_test, y_pred_test)
print(f"Testing Accuracy: {rf_accuracy:.2%}")
```

```python
# get feature importances
importances = optimized_rf_classifier.feature_importances_

# get the indices of features in descending order of importance
indices = np.argsort(importances)[::-1] # return the indices that would sort the array, and reverse the array to have the indices from most to least important

feature_names = X_train.columns

# print the feature ranking
print("Feature ranking:")
for rank, index in enumerate(indices, 1):  # (index, element)
    feature_name = feature_names[index]
    importance_value = importances[index]
    print(f"{rank}. Feature {index}: {feature_name} ({importance_value:.4f})")
# eg. the largest element has the lowest rank

# plot the feature importances
# X_train.shape[1] represents the number of features in the dataset
plt.figure(figsize=(10, 6))
plt.title("Feature Importances")
plt.bar(range(X_train.shape[1]), importances[indices], align = "center")
plt.xticks(range(X_train.shape[1]), indices)
plt.xlabel("Feature Index")
plt.ylabel("Importance")
plt.show()
```

```python

user_input = {}

yes_no_params = ['Mood Swing', 'Suicidal thoughts', 'Anorexia', 'Authority Respect', 'Try-Explanation',
                 'Aggressive Response', 'Ignore & Move-On', 'Nervous Break-down', 'Admit Mistakes', 'Overthinking']

number_params_extracted = ['Sexual Activity', 'Concentration', 'Optimism']

number_params_frequency = ['Sadness', 'Euphoric', 'Exhausted', 'Sleep dissorder']

print("Please provide your answers for the following questions:")


for param in yes_no_params:
    while True:
        response = input(f"Do you experience '{param}'? (yes/no): ").strip().lower()
        if response in ['yes', 'no']:
            user_input[param] = 1 if response == 'yes' else 0
            break
        else:
            print("Invalid input. Please enter 'yes' or 'no'.")

for param in number_params_extracted:
    while True:
        try:
            response = int(input(f"On a scale of 1-5, how would you rate your '{param}'? (Enter a number): "))
            if 1 <= response <= 5:
                user_input[param] = response
                break
            else:
                print("Invalid input. Please enter a number between 1 and 5.")
        except ValueError:
            print("Invalid input. Please enter a number.")


frequency_mapping = {'seldom': 0, 'sometimes': 1, 'usually': 2, 'most-often': 3}
reverse_frequency_mapping = {0: 'Seldom', 1: 'Sometimes', 2: 'Usually', 3: 'Most-Often'}

for param in number_params_frequency:
    while True:
        response = input(f"How often do you experience '{param}'? (seldom, sometimes, usually, most-often): ").strip().lower()
        if response in frequency_mapping:
            user_input[param] = frequency_mapping[response]
            break
        else:
            print("Invalid input. Please enter one of: seldom, sometimes, usually, most-often.")


user_df = pd.DataFrame([user_input])

input_features = X.columns.tolist()
user_df = user_df[input_features]

best_model_name = models.sort_values(by='Score', ascending=False).iloc[0]['Model']

print(f"\nThe best performing model is: {best_model_name}")

predicted_diagnose_encoded = optimized_rf_classifier.predict(user_df)


reverse_mapping_dict = {0: 'Normal', 1: 'Bipolar Type-1', 2: 'Bipolar Type-2', 3: 'Depression'}
predicted_diagnose = reverse_mapping_dict[predicted_diagnose_encoded[0]]

print(f"\nBased on your responses, the predicted mental disorder is: {predicted_diagnose}")

```

```python

!pip install gradio

import gradio as gr
import pandas as pd
import numpy as np
import joblib
joblib.dump(optimized_rf_classifier, 'random_forest_model.pkl')

def predict_mental_disorder(mood_swing, suicidal_thoughts, anorexia, authority_respect, try_explanation,
                            aggressive_response, ignore_move_on, nervous_break_down, admit_mistakes,
                            overthinking, sexual_activity, concentration, optimism, sadness, euphoric,
                            exhausted, sleep_disorder):


    user_input = {
        'Mood Swing': 1 if mood_swing == 'Yes' else 0,
        'Suicidal thoughts': 1 if suicidal_thoughts == 'Yes' else 0,
        'Anorexia': 1 if anorexia == 'Yes' else 0,
        'Authority Respect': 1 if authority_respect == 'Yes' else 0,
        'Try-Explanation': 1 if try_explanation == 'Yes' else 0,
        'Aggressive Response': 1 if aggressive_response == 'Yes' else 0,
        'Ignore & Move-On': 1 if ignore_move_on == 'Yes' else 0,
        'Nervous Break-down': 1 if nervous_break_down == 'Yes' else 0,
        'Admit Mistakes': 1 if admit_mistakes == 'Yes' else 0,
        'Overthinking': 1 if overthinking == 'Yes' else 0,
        'Sexual Activity': sexual_activity,
        'Concentration': concentration,
        'Optimism': optimism,
        'Sadness': {'Seldom': 0, 'Sometimes': 1, 'Usually': 2, 'Most-Often': 3}[sadness],
        'Euphoric': {'Seldom': 0, 'Sometimes': 1, 'Usually': 2, 'Most-Often': 3}[euphoric],
        'Exhausted': {'Seldom': 0, 'Sometimes': 1, 'Usually': 2, 'Most-Often': 3}[exhausted],
        'Sleep dissorder': {'Seldom': 0, 'Sometimes': 1, 'Usually': 2, 'Most-Often': 3}[sleep_disorder]
    }


    user_df = pd.DataFrame([user_input])

    try:
        input_features = X.columns.tolist()
        user_df = user_df[input_features]
    except NameError:
        print("Error: 'X' not found. Please ensure the DataFrame 'X' is defined before running this cell.")
        return "Prediction failed: Missing training data information."



    loaded_model = joblib.load('random_forest_model.pkl')


    predicted_diagnose_encoded = loaded_model.predict(user_df)


    reverse_mapping_dict = {0: 'Normal', 1: 'Bipolar Type-1', 2: 'Bipolar Type-2', 3: 'Depression'}
    predicted_diagnose = reverse_mapping_dict[predicted_diagnose_encoded[0]]

    return predicted_diagnose


yes_no_choices = ['Yes', 'No']
frequency_choices = ['Seldom', 'Sometimes', 'Usually', 'Most-Often']

inputs = [
    gr.Radio(yes_no_choices, label="Mood Swing?"),
    gr.Radio(yes_no_choices, label="Suicidal thoughts?"),
    gr.Radio(yes_no_choices, label="Anorexia?"),
    gr.Radio(yes_no_choices, label="Authority Respect?"),
    gr.Radio(yes_no_choices, label="Try-Explanation?"),
    gr.Radio(yes_no_choices, label="Aggressive Response?"),
    gr.Radio(yes_no_choices, label="Ignore & Move-On?"),
    gr.Radio(yes_no_choices, label="Nervous Break-down?"),
    gr.Radio(yes_no_choices, label="Admit Mistakes?"),
    gr.Radio(yes_no_choices, label="Overthinking?"),

    gr.Slider(minimum=1, maximum=5, step=1, label="Sexual Activity (1-5)?"),
    gr.Slider(minimum=1, maximum=5, step=1, label="Concentration (1-5)?"),
    gr.Slider(minimum=1, maximum=5, step=1, label="Optimism (1-5)?"),
    gr.Radio(frequency_choices, label="How often do you experience Sadness?"),
    gr.Radio(frequency_choices, label="How often do you experience Euphoric?"),
    gr.Radio(frequency_choices, label="How often do you experience Exhausted?"),
    gr.Radio(frequency_choices, label="How often do you experience Sleep disorder?")
]

output = gr.Textbox(label="Predicted Mental Disorder")


gr.Interface(fn=predict_mental_disorder, inputs=inputs, outputs=output, title="Mental Disorder Prediction").launch(share=True)
```
